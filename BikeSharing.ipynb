```python
Labels = ['Spring', 'Summer', 'Autumn', 'Winter']
Counts = spring['Count'].sum(), summer['Count'].sum(), autumn['Count'].sum(), winter['Count'].sum()
```

```python
figure2, axes2 = plt.subplots(figsize=(14,8))

axes2.pie(Counts, labels = Labels, autopct = '%1.1f%%', explode= [0.1,0,0.1,0])

axes2.legend()

plt.show()
```
